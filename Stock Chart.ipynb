# The purpose of this file is to display stock information.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas_datareader as web
from datetime import datetime, timedelta
```

```python
for stock in ['AAPL', 'NFLX', 'AMZN', 'TSLA', 'FB', 'GOOG', 'MSFT']:

    #Getting today and the past 60 days dates to grab data for
    today = datetime.now()
    past = today - timedelta(days = 60)
    today = today.strftime("%Y-%m-%d")
    past = past.strftime("%Y-%m-%d")


    #Getting our data
    df = web.DataReader(stock, data_source = 'yahoo', start = past, end = today)

    #Visualizing the closing price history
    sns.set_style('whitegrid')
    plt.figure(figsize = (16,10))
    sns.lineplot(data = df[['Close']])


    plt.ylabel('Closing Price (USD)', fontsize = 18)
    plt.yticks(fontsize = 18)
    plt.xlabel('Date', fontsize = 18)
    plt.xticks(rotation = 45, fontsize = 18)


    plt.title(stock + ' Closing Price (USD) - Past 60 Days', fontsize = 18)
    plt.legend('')
    plt.savefig('static/images/' + stock + '.png')
    plt.show()
```
